```python
parser = CandidateParser()
career_extractor = CareerFeatureExtractor()
engine = EmbeddingEngine()
```

```python
import pandas as pd
df = pd.read_json(
    r"D:\India Runs\TalentDNA_AI\data\raw\candidates.jsonl",
    lines=True
)
sample_df = df.head(100)
```

```python
candidate_embeddings = []
candidate_ids = []
```

```python
for _, row in sample_df.iterrows():

    parsed = parser.parse(row)

    career = career_extractor.extract(
        row["career_history"]
    )

    candidate = {
        **parsed,
        **career
    }

    emb = engine.encode_candidate(
        candidate
    )

    candidate_embeddings.append(
        emb
    )

    candidate_ids.append(
        candidate["candidate_id"]
    )
```

```python
from src.retrieval import (
    CandidateRetriever
)

retriever = CandidateRetriever()

retriever.build_index(
    candidate_embeddings,
    candidate_ids
)
```
