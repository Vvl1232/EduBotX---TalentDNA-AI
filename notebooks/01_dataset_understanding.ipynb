```python
signal_values = {}

for s in df["redrob_signals"]:
    for k, v in s.items():

        if k not in signal_values:
            signal_values[k] = set()

        if isinstance(v, dict):
            signal_values[k].add(str(v))
        else:
            signal_values[k].add(v)

for k, vals in signal_values.items():
    print(f"\n{k}")
    print(list(vals)[:5])  # sample values
```
